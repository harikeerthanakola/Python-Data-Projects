```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("netflix_titles.csv")
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
# Convert date_added to datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Fill missing values
df['director'].fillna("Unknown", inplace=True)
df['cast'].fillna("Unknown", inplace=True)
df['country'].fillna("Unknown", inplace=True)

# Drop rows with missing rating (optional)
df.dropna(subset=['rating'], inplace=True)
```

```python
# Extract year added
df['year_added'] = df['date_added'].dt.year

# Create decade column
df['release_decade'] = (df['release_year'] // 10) * 10

# Create is_movie column
df['is_movie'] = df['type'].apply(lambda x: 1 if x == "Movie" else 0)
```

```python
type_counts = df['type'].value_counts()

plt.figure()
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%')
plt.title("Movies vs TV Shows")
plt.show()
```

```python
plt.figure()
df['year_added'].value_counts().sort_index().plot(kind='line')
plt.title("Content Added Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.show()
```

```python
# Split genres
genres = df['listed_in'].str.split(', ', expand=True).stack()

plt.figure()
genres.value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Genres on Netflix")
plt.xticks(rotation=45)
plt.show()
```

```python
countries = df['country'].str.split(', ', expand=True).stack()

plt.figure()
countries.value_counts().head(10).plot(kind='bar')
plt.title("Top Countries Producing Content")
plt.xticks(rotation=45)
plt.show()
```

```python
pivot = pd.crosstab(df['year_added'], df['type'])

plt.figure()
sns.heatmap(pivot, cmap="coolwarm")
plt.title("Heatmap of Content Type Over Years")
plt.show()
```

```python
df.to_csv("cleaned_data.csv", index=False)
```

## Key Insights

- Netflix content increased rapidly after 2015
- Movies dominate over TV shows
- USA and India produce the most content
- Drama and Comedy are the most popular genres
